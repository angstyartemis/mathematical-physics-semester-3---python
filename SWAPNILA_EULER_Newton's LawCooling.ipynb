### Solving the differential equation of Newton's law of Cooling using Euler's Algorithm

Newton's Law of Cooling states that the rate of heat loss of a body is directly propotional to the difference of temperatures between it and its surrounding.
The differential equation of the same is as follows:
$$ \frac{dT}{dt}= -k(T-T_o) [here, T_o is the surrounding temperature] $$

Now,we are going to solve the equation using Euler's algorithm:

#### Step 1: DIMENSIONAL ANALYSIS.

The equation has a natural time scale $t_o = 1/k$ .We define a dimensionless quantity $\tau$ which is equal to the ratio between the 2 time-scales($\tau= t/t_o$). Now, we re-write the equation with respect to $\tau$. $$ \frac{dT}{d\tau}= -(T-T_o)$$
The temperature of the body being $T_a$ at the beginning, this method helps us determine $T(\tau)$.
We divide the interval $[0,\tau]$ into $M$ equal intervals where $\Delta\tau=\tau/M$.
Using Euler's approximation, $$ T(\tau+\Delta\tau) = T(\tau)+\Delta\tau\times\frac {dt}{d\tau}$$
which is equal to, $$T(\tau+\Delta\tau)=T(\tau)-\Delta\tau\times(T-T_o)$$
Now, we can assign a variable to the $(T-T_o)$ and use the above equation iteratively to compute temperature.












```python
#we are solving for a particular case with initial temp 318K and surrounding temp.303K
temp= 318 #initialtemperatureofbody
temp_sur= 303# temperatureofsurrounding
dif=temp-temp_sur
tau=0
tau_final=5 #moment till which computation is done
delta_tau=0.1
N_slices= int(tau_final/delta_tau)
taus=[tau]
temps=[temp]
```

```python
for step in range(0,N_slices):
    temp+= -(delta_tau*dif)
    dif= temp-temp_sur
    tau+=delta_tau
    temps.append(temp)
    taus.append(tau)
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel('tau')
plt.ylabel('temperature')
plt.style.use('seaborn-whitegrid')
plt.title("Newton's Law of Cooling")
plt.plot(taus,temps);
```

Now, we plot the exact analytical solution of this differential equation 
 $$T(\tau)= T_o+ C e^{-\tau}$$
 For our problem, $C=15$.
 Now, we plot the correct graph.
 

```python
import math 
def exp(t):
  return math.exp(-t)
    
```

```python
correct_temps=[]
for i in taus:
    correct_temps.append((temp_sur+(15*exp(i))))
```

```python
plt.xlabel('tau')
plt.ylabel('temp')
plt.style.use('seaborn-whitegrid')
plt.title('comparison of graphs')
plt.plot(taus,temps,'-b',label='euler solution')
plt.plot(taus,correct_temps,'-r',label='analytical solution')
plt.legend();
```
