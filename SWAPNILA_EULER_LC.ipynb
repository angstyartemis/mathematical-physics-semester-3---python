```python
import math
import matplotlib.pyplot as plt # Plotting library imported
%matplotlib inline 
```

```python
current_fraction = 0
tau = 0
tau_final = 10
delta_tau = 0.1
divisions = int(tau_final/delta_tau)
current_fractions = [current_fraction]
taus = [tau]
```

```python
for i in range (0, divisions):
    current_fraction += (delta_tau*(1-current_fraction))
    tau += delta_tau
    current_fractions.append(current_fraction)
    taus.append(tau)
```

```python
plt.xlabel('tau')
plt.ylabel('Q')
plt.style.use('seaborn-whitegrid')
plt.title('charging in LC circuit')
plt.plot(taus, current_fractions);
```

```python
import matplotlib.pyplot as plt
import math
def correct_current_fraction (t):
    return 1- math.exp(-t)
```

```python
correct_current_fractions= []
for i in taus:
    correct_current_fractions.append(correct_current_fraction(i))
```
