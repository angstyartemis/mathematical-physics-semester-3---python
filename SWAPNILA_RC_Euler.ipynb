```python
import math
import matplotlib.pyplot as plt # Plotting library imported
%matplotlib inline 
```

```python
charge_fraction = 0
tau = 0
tau_final = 10
delta_tau = 0.1
divisions = int(tau_final/delta_tau)
charge_fractions = [charge_fraction]
taus = [tau]
```

```python
for i in range (0, divisions):
    charge_fraction += (delta_tau*(1-charge_fraction))
    tau += delta_tau
    charge_fractions.append(charge_fraction)
    taus.append(tau)
```

```python
plt.xlabel('tau')
plt.ylabel('Q')
plt.style.use('seaborn-whitegrid')
plt.title('charging of capacitor in RC circuit')
plt.plot(taus, charge_fractions);
```

```python
import matplotlib.pyplot as plt
import math
def correct_charge_fraction (t):
    return 1- math.exp(-t)
```

```python
correct_charge_fractions= []
for i in taus:
    correct_charge_fractions.append(correct_charge_fraction(i))
```

```python
plt.xlabel('taus')
plt.ylabel('Q')
plt.style.use('seaborn-whitegrid')
plt.title('charging the capacitor in RC circuit')
plt.plot(taus,charge_fractions, '-b', label='charge_fraction')
plt.plot(taus,correct_charge_fractions, '-r', label='correct_charge_fraction')
plt.legend();
```
