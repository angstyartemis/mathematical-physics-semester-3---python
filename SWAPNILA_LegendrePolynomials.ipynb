```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#generatingFunctionforLegendrePolynomial
def Legendre(n, x):
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        return ((2 * n - 1) / n) * x * Legendre(n-1, x) - ((n - 1) / n) * Legendre(n-2, x)
```

```python
n = 4
x = [i/100 for i in range(-100, 101)]
y = [Legendre(n, xi) for xi in x]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("P{}(x)".format(n))
```

```python
ns = [0, 1, 2, 3, 4, 5, 6]
for n in ns:
    y = [Legendre(n, xi) for xi in x]
    plt.plot(x, y, label="P{}".format(n))
plt.xlabel("x")
plt.ylabel("P(x)")
plt.title("Plot of first {} Legendre Polynomials".format(len(ns)))
plt.legend()
```
