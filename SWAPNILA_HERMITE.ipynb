### HermitePolynomials

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def factorial(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x*factorial(x-1)
```



RECURSIVE FUNCTION TO GENERATE HERMITE POLYNOMIAL OF ORDER n AT x

```python
def Hermite(n, x):
    if n == 0:
        return 1
    elif n == 1:
        return 2*x
    else:
        return Hermite(n-1,x)*2*x - Hermite(n-2,x)*(n-1)*2
```

```python
x = [i/100 for i in range(-100, 101)]
ns = [0, 1, 2, 3, 4]
for n in ns:
    y = [Hermite(n, xi) for xi in x]
    plt.plot(x, y, label="H{}".format(n))
plt.xlabel("x")
plt.ylabel("H(x)")
plt.title("Plot of first {} Hermite Polynomials".format(len(ns)))
plt.legend()

```

```python
# COMPUTING SIMPSON'S INTEGRAL

def func(T):
    return (Hermite(2,T)**2)*math.exp(-(T)**2)

       
```

```python
a=-10
b=10
M=25
N=2*M
delta= (b-a)/N
sum_even=0
oddsum= func(a+delta)
```

```python
for i in range (1, M):
    sum_even+= func(a + 2*i*delta)
    oddsum+= func(a + (2*i+1)*delta)
    
I= ( func(a)+func(b)+4*oddsum +2*sum_even)*delta/3
print(I)
```

```python
#plotting given hermittian relation
def g(n,a):
        A = Hermite(n,a)*math.exp(-a*a/2)
        B= math.pow((math.pow(math.pi,0.5))*factorial(n)*math.pow(2,n),-0.5)
        return A*B
for n in ns:
    y = [g(n, xi) for xi in x]
    plt.plot(x, y, label="H{}".format(n))
plt.xlabel("x")
plt.ylabel("H(x)")
plt.title("Plot of first {} Hermite Polynomials orthogonality Relations".format(len(ns)))
plt.legend()       
```
