### Plotting the graphs of Newton's Laws of motion for pendulum with drag co-efficient in velocity

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline

theta=0.1 #initialangle
vel=0 #initial velocity
E=(0.5*vel**2)+4*math.pi**2*(1-math.cos(theta))
tau=0
tau_final=10
delta_tau=0.001
N_slices= int(tau_final/delta_tau)
thetas=[theta]
vels=[vel]
Es=[E]
taus=[tau]
```

```python
plt.xlabel('tau')
plt.plot(taus,thetas, linestyle='-', color='red', label='position')
plt.plot(taus,vels, linestyle='-',color='blue', label='velocity')
plt.legend();


```

```python
plt.plot(taus,Es);
```

```python
b=1
for i in range (N_slices):
    theta_t=theta +vel*delta_tau
    vel_t= vel-4*math.pi**2*delta_tau*math.sin(theta)-b*vel*delta_tau
    theta=theta_t
    vel=vel_t
    E=(0.5*vel**2) + 4*math.pi**2*(1-math.cos(theta))
    tau+=delta_tau
    thetas.append(theta)
    vels.append(vel)
    Es.append(E)
    taus.append(tau)
plt.xlabel('tau')
plt.plot(taus,thetas, linestyle='-', color='red', label='position')
plt.plot(taus,vels, linestyle='-',color='blue', label='velocity')
plt.legend();
   
    
    
    
    
```
